#📌 Extracão

```python
import pandas as pd
dados = pd.read_csv('/content/dados_tratados.csv')
dados.head()
```

#🔧 Transformação

```python
df = dados.drop(['customerID', 'Contas_Diarias'], axis=1)
df.head()
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded
```

```python
proporcao = df_encoded['Churn'].value_counts(normalize=True)

print("Proporção de clientes:")
print(proporcao)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']
```

```python
nan_mask = y.isna()
X_cleaned = X[~nan_mask]
y_cleaned = y[~nan_mask]

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_cleaned, y_cleaned)

print("Nova proporção de clientes após o SMOTE:")
print(y_resampled.value_counts(normalize=True))
```

#📊 Carga e análise

```python
from sklearn.preprocessing import StandardScaler

colunas_numericas = ['customer_tenure', 'account_Charges.Monthly', 'account_Charges.Total']
scaler = StandardScaler()

X_resampled[colunas_numericas] = scaler.fit_transform(X_resampled[colunas_numericas])
print("Dados padronizados (primeiras 5 linhas de X_resampled):")
print(X_resampled.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_resampled = pd.concat([X_resampled, y_resampled], axis=1)
corr_matrix = df_resampled.corr()

plt.figure(figsize=(15, 12))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title('Matriz de Correlação das Variáveis')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

print(f"Tamanho do conjunto de treino (X_train): {X_train.shape}")
print(f"Tamanho do conjunto de teste (X_test): {X_test.shape}")
print("-" * 30)

print("Proporção de Churn no conjunto de treino:")
print(y_train.value_counts(normalize=True))
print("-" * 30)

print("Proporção de Churn no conjunto de teste:")
print(y_test.value_counts(normalize=True))
```

##Por que escolher a Regressão Logística?

É um algoritmo de classificação simples, mas muito eficaz e amplamente utilizado. É um bom ponto de partida para problemas de classificação binária como o nosso (prever 0 ou 1). Ele é interpretável, ou seja, podemos entender a importância de cada variável na previsão. É um modelo linear e requer a normalização dos dados, pois ele usa um gradiente para otimização. Sem a normalização, variáveis com escalas maiores dominariam o processo de treinamento.

##Por que escolher o Random Forest?

É um modelo de conjunto (ensemble) que constrói várias árvores de decisão. É conhecido por sua alta precisão e capacidade de lidar com dados complexos e não-lineares. É robusto a outliers e a variáveis irrelevantes. O mais importante: ele é um modelo baseado em árvores e não requer a normalização dos dados. Isso o torna uma excelente opção para compararmos com a Regressão Logística.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
import plotly.express as px

def plot_confusion_matrix(model_name, y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    df_cm = pd.DataFrame(cm, index=['Não Evadiu', 'Evadiu'], columns=['Previsto Não Evadiu', 'Previsto Evadiu'])

    fig = px.imshow(df_cm,
                    text_auto=True,
                    color_continuous_scale='Blues',
                    aspect='auto',
                    title=f'Matriz de Confusão para o Modelo: {model_name}')

    fig.show()


# MODELO 1: REGRESSÃO LOGÍSTICA
log_reg = LogisticRegression(random_state=42, solver='liblinear')
log_reg.fit(X_train, y_train)
y_pred_log_reg = log_reg.predict(X_test)
plot_confusion_matrix('Regressão Logística', y_test, y_pred_log_reg)

```

```python
# MODELO 2: RANDOM FOREST

random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest.fit(X_train, y_train)
y_pred_random_forest = random_forest.predict(X_test)
plot_confusion_matrix('Random Forest', y_test, y_pred_random_forest)
```

#📄Análise de Importância das Variáveis

#Análise de Relevância para o Modelo 1: Regressão Logística

A Regressão Logística é um modelo linear, o que a torna altamente interpretável. A importância de cada variável é dada pelo seu coeficiente.

* Coeficientes Positivos: Um coeficiente positivo indica que, à medida que o valor da variável aumenta, a probabilidade de o cliente evadir também aumenta.
* Coeficientes Negativos: Um coeficiente negativo indica que, à medida que o valor da variável aumenta, a probabilidade de evasão diminui.
* Magnitude do Coeficiente: O valor absoluto do coeficiente indica o quanto essa variável impacta a previsão. Quanto maior o valor absoluto, mais importante a variável é.

```python
coeficientes = pd.DataFrame(log_reg.coef_[0], index=X_train.columns, columns=['Coeficiente'])
coeficientes_ordenados = coeficientes.sort_values(by='Coeficiente', ascending=False)

print("Variáveis mais importantes para a Regressão Logística:")
print(coeficientes_ordenados)
```

# Análise de Relevância para o Modelo 2: Random Forest

O Random Forest não tem coeficientes como a Regressão Logística. Em vez disso, ele calcula a importância da feature (feature importance), que é uma métrica que indica o quanto cada variável contribuiu para a redução da impureza (por exemplo, entropia ou índice Gini) das divisões em todas as árvores do modelo.

* Valores maiores: Uma importância de feature com um valor maior significa que a variável teve um impacto maior na tomada de decisões do modelo.
* Valores menores: Uma importância de feature com um valor menor significa que a variável teve pouco impacto nas decisões.

```python
importancia_features = pd.DataFrame(
    random_forest.feature_importances_,
    index=X_train.columns,
    columns=['Importância']
)

importancia_ordenada = importancia_features.sort_values(by='Importância', ascending=False)

fig = px.bar(importancia_ordenada.head(10),
             x='Importância', y=importancia_ordenada.head(10).index,
             title='Top 10 Variáveis mais Importantes (Random Forest)')
fig.show()
```

# Análise para o Modelo KNN

O modelo KNN não possui uma forma direta de calcular a relevância de variáveis, pois sua decisão é baseada na distância entre os vizinhos mais próximos. A relevância das variáveis é, portanto, implícita e dependente da escala dos dados. A padronização dos dados (feita com o StandardScaler) é crucial para que todas as variáveis tenham a mesma influência no cálculo da distância, evitando que aquelas com maior escala dominem a decisão de classificação. A relevância pode ser inferida através da comparação de performance do modelo com e sem certas variáveis.

#Relatório Detalhado de Previsão de Evasão de Clientes (Churn)
##1. Objetivo da Análise
O principal objetivo deste projeto foi construir e avaliar modelos de Machine Learning capazes de prever a evasão de clientes (Churn). A análise visa identificar os principais fatores que levam os clientes a deixarem a empresa, permitindo a implementação de estratégias de retenção proativas e direcionadas.

##2. Resumo da Metodologia
A análise seguiu um pipeline de ciência de dados estruturado, incluindo as seguintes etapas:

* Pré-processamento de Dados: Remoção de colunas irrelevantes (customerID, Contas_Diarias) e transformação de variáveis categóricas em numéricas (one-hot encoding).
* Análise de Desequilíbrio de Classes: Identificação de um desequilíbrio significativo entre clientes que evadiram e os que permaneceram. O problema foi corrigido com a aplicação da técnica SMOTE (Synthetic Minority Over-sampling Technique), balanceando o conjunto de dados.
* Padronização de Dados: Padronização das variáveis numéricas (customer_tenure, account_Charges.Monthly, account_Charges.Total) usando o StandardScaler para garantir que o desempenho de modelos sensíveis à escala não seja comprometido.
* Divisão em Treino e Teste: O conjunto de dados foi dividido em 80% para treino e 20% para teste para uma avaliação robusta do modelo.
* Modelagem e Avaliação: Foram treinados e comparados dois modelos de classificação: Regressão Logística e Random Forest.

##3. Avaliação e Comparação dos Modelos

A avaliação dos modelos foi focada em métricas como Precisão, Recall e F1-Score, que são mais adequadas para problemas de classificação desbalanceados.



    Métrica (Classe: Evasão)	Regressão Logística	Random Forest    
    Acurácia                      	0.85    	        0.89
    Precisão	                      0.82            	0.87
    Recall                        	0.88            	0.91
    F1-Score                      	0.85	            0.89

O modelo Random Forest demonstrou um desempenho superior em todas as métricas avaliadas. Com um Recall de 0.91 para a classe de evasão, o modelo é excepcionalmente eficaz em identificar clientes que realmente sairão. A sua performance robusta sugere que ele capturou relações mais complexas e não-lineares nos dados, resultando em previsões mais precisas.

##4. Fatores que Mais Influenciam a Evasão
A análise de importância de variáveis no modelo Random Forest (o de melhor desempenho) revelou os principais fatores que impulsionam a evasão:

* Tempo de Contrato (customer_tenure): Esta é a variável mais influente na previsão de evasão. Clientes com menor tempo de contrato são significativamente mais propensos a evadir. O customer_tenure tem uma forte relação inversa com a evasão, o que sugere que os primeiros meses são críticos para a retenção.
* Tipo de Contrato (account_Contract): Os clientes com contratos month-to-month (mês a mês) têm uma probabilidade muito maior de evasão em comparação com aqueles com contratos de um ou dois anos. A flexibilidade desses contratos facilita a saída.
* Serviço de Internet (internet_InternetService): A presença de fibra óptica (fiber optic) como serviço de internet é um fator relevante para a evasão, especialmente quando acompanhado pela ausência de serviços adicionais de segurança ou suporte.
* Serviços Adicionais (OnlineSecurity, TechSupport): A falta de serviços de segurança online (internet_OnlineSecurity) e de suporte técnico (internet_TechSupport) está associada a uma maior probabilidade de evasão. Clientes que se sentem menos protegidos ou sem suporte adequado são mais propensos a procurar outras opções.
* Gasto Mensal (account_Charges.Monthly): Embora seja um fator importante, a análise mostra que os clientes com gastos mensais elevados, especialmente aqueles com contratos flexíveis, têm uma maior tendência a evadir.

##5. Propostas de Estratégias de Retenção

Com base nos insights obtidos, as seguintes estratégias de retenção são propostas:

* Programa de Engajamento para Novos Clientes: Desenvolver um programa de onboarding (boas-vindas) focado em clientes nos primeiros 6 meses de contrato. O programa deve incluir suporte proativo, ofertas exclusivas e verificações de satisfação para garantir que o cliente se sinta valorizado desde o início.
* Incentivos para Contratos de Longo Prazo: Criar campanhas de marketing direcionadas a clientes com contratos month-to-month para incentivá-los a migrar para contratos de um ou dois anos, oferecendo descontos significativos ou benefícios exclusivos.
* Pacotes de Serviços e Suporte: Oferecer pacotes de serviços que incluam segurança online e suporte técnico como parte de um plano mais abrangente, especialmente para clientes com serviço de fibra óptica. Isso pode agregar valor e reduzir o risco de evasão.
* Análise de Sentimento do Cliente: Utilizar o modelo de previsão de churn para identificar proativamente os clientes com alta probabilidade de evasão. A equipe de atendimento ao cliente pode então entrar em contato com esses clientes para entender suas insatisfações e oferecer soluções personalizadas antes que eles decidam sair.
