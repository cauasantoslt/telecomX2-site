```python
import pandas as pd
dados = pd.read_csv('/content/dados_tratados.csv')
dados.head()
```

#ðŸ”§ TransformaÃ§Ã£o

```python
df = dados.drop(['customerID', 'Contas_Diarias'], axis=1)
df.head()
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded
```

```python
proporcao = df_encoded['Churn'].value_counts(normalize=True)

print("ProporÃ§Ã£o de clientes:")
print(proporcao)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']
```

```python
nan_mask = y.isna()
X_cleaned = X[~nan_mask]
y_cleaned = y[~nan_mask]

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_cleaned, y_cleaned)

print("Nova proporÃ§Ã£o de clientes apÃ³s o SMOTE:")
print(y_resampled.value_counts(normalize=True))
```

#ðŸ“Š Carga e anÃ¡lise

```python
from sklearn.preprocessing import StandardScaler

colunas_numericas = ['customer_tenure', 'account_Charges.Monthly', 'account_Charges.Total']
scaler = StandardScaler()

X_resampled[colunas_numericas] = scaler.fit_transform(X_resampled[colunas_numericas])
print("Dados padronizados (primeiras 5 linhas de X_resampled):")
print(X_resampled.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_resampled = pd.concat([X_resampled, y_resampled], axis=1)
corr_matrix = df_resampled.corr()

plt.figure(figsize=(15, 12))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title('Matriz de CorrelaÃ§Ã£o das VariÃ¡veis')
plt.show()
```

```python
import plotly.express as px
import pandas as pd

df_resampled = pd.concat([X_resampled, y_resampled], axis=1)
corr_matrix = df_resampled.corr()

fig = px.imshow(corr_matrix,
                labels=dict(x="VariÃ¡veis", y="VariÃ¡veis", color="CorrelaÃ§Ã£o"),
                x=corr_matrix.columns,
                y=corr_matrix.columns,
                color_continuous_scale='RdBu_r')

fig.update_layout(title_text='Matriz de CorrelaÃ§Ã£o Interativa',
                  title_x=0.5,
                  xaxis_showgrid=False,
                  yaxis_showgrid=False,
                  yaxis_autorange='reversed')

fig.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

print(f"Tamanho do conjunto de treino (X_train): {X_train.shape}")
print(f"Tamanho do conjunto de teste (X_test): {X_test.shape}")
print("-" * 30)

print("ProporÃ§Ã£o de Churn no conjunto de treino:")
print(y_train.value_counts(normalize=True))
print("-" * 30)

print("ProporÃ§Ã£o de Churn no conjunto de teste:")
print(y_test.value_counts(normalize=True))
```

##Por que escolher a RegressÃ£o LogÃ­stica?

Ã‰ um algoritmo de classificaÃ§Ã£o simples, mas muito eficaz e amplamente utilizado. Ã‰ um bom ponto de partida para problemas de classificaÃ§Ã£o binÃ¡ria como o nosso (prever 0 ou 1). Ele Ã© interpretÃ¡vel, ou seja, podemos entender a importÃ¢ncia de cada variÃ¡vel na previsÃ£o. Ã‰ um modelo linear e requer a normalizaÃ§Ã£o dos dados, pois ele usa um gradiente para otimizaÃ§Ã£o. Sem a normalizaÃ§Ã£o, variÃ¡veis com escalas maiores dominariam o processo de treinamento.

##Por que escolher o Random Forest?

Ã‰ um modelo de conjunto (ensemble) que constrÃ³i vÃ¡rias Ã¡rvores de decisÃ£o. Ã‰ conhecido por sua alta precisÃ£o e capacidade de lidar com dados complexos e nÃ£o-lineares. Ã‰ robusto a outliers e a variÃ¡veis irrelevantes. O mais importante: ele Ã© um modelo baseado em Ã¡rvores e nÃ£o requer a normalizaÃ§Ã£o dos dados. Isso o torna uma excelente opÃ§Ã£o para compararmos com a RegressÃ£o LogÃ­stica.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
import plotly.express as px

def plot_confusion_matrix(model_name, y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    df_cm = pd.DataFrame(cm, index=['NÃ£o Evadiu', 'Evadiu'], columns=['Previsto NÃ£o Evadiu', 'Previsto Evadiu'])

    fig = px.imshow(df_cm,
                    text_auto=True,
                    color_continuous_scale='Blues',
                    aspect='auto',
                    title=f'Matriz de ConfusÃ£o para o Modelo: {model_name}')

    fig.show()


# MODELO 1: REGRESSÃƒO LOGÃSTICA
log_reg = LogisticRegression(random_state=42, solver='liblinear')
log_reg.fit(X_train, y_train)
y_pred_log_reg = log_reg.predict(X_test)
plot_confusion_matrix('RegressÃ£o LogÃ­stica', y_test, y_pred_log_reg)

```

```python
# MODELO 2: RANDOM FOREST

random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest.fit(X_train, y_train)
y_pred_random_forest = random_forest.predict(X_test)
plot_confusion_matrix('Random Forest', y_test, y_pred_random_forest)
```

#AnÃ¡lise de RelevÃ¢ncia para o Modelo 1: RegressÃ£o LogÃ­stica

A RegressÃ£o LogÃ­stica Ã© um modelo linear, o que a torna altamente interpretÃ¡vel. A importÃ¢ncia de cada variÃ¡vel Ã© dada pelo seu coeficiente.

* Coeficientes Positivos: Um coeficiente positivo indica que, Ã  medida que o valor da variÃ¡vel aumenta, a probabilidade de o cliente evadir tambÃ©m aumenta.
* Coeficientes Negativos: Um coeficiente negativo indica que, Ã  medida que o valor da variÃ¡vel aumenta, a probabilidade de evasÃ£o diminui.
* Magnitude do Coeficiente: O valor absoluto do coeficiente indica o quanto essa variÃ¡vel impacta a previsÃ£o. Quanto maior o valor absoluto, mais importante a variÃ¡vel Ã©.

```python
coeficientes = pd.DataFrame(log_reg.coef_[0], index=X_train.columns, columns=['Coeficiente'])
coeficientes_ordenados = coeficientes.sort_values(by='Coeficiente', ascending=False)

print("VariÃ¡veis mais importantes para a RegressÃ£o LogÃ­stica:")
print(coeficientes_ordenados)
```

# AnÃ¡lise de RelevÃ¢ncia para o Modelo 2: Random Forest

O Random Forest nÃ£o tem coeficientes como a RegressÃ£o LogÃ­stica. Em vez disso, ele calcula a importÃ¢ncia da feature (feature importance), que Ã© uma mÃ©trica que indica o quanto cada variÃ¡vel contribuiu para a reduÃ§Ã£o da impureza (por exemplo, entropia ou Ã­ndice Gini) das divisÃµes em todas as Ã¡rvores do modelo.

* Valores maiores: Uma importÃ¢ncia de feature com um valor maior significa que a variÃ¡vel teve um impacto maior na tomada de decisÃµes do modelo.
* Valores menores: Uma importÃ¢ncia de feature com um valor menor significa que a variÃ¡vel teve pouco impacto nas decisÃµes.

```python
importancia_features = pd.DataFrame(
    random_forest.feature_importances_,
    index=X_train.columns,
    columns=['ImportÃ¢ncia']
)

importancia_ordenada = importancia_features.sort_values(by='ImportÃ¢ncia', ascending=False)

fig = px.bar(importancia_ordenada.head(10),
             x='ImportÃ¢ncia', y=importancia_ordenada.head(10).index,
             title='Top 10 VariÃ¡veis mais Importantes (Random Forest)')
fig.show()
```
